```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/kaggle/input/customer-personality-analysis/marketing_campaign.csv", sep='\t')
```

```python
# 1. Check missing values
print("Missing values:\n", df.isnull().sum())
```

```python
# Handle missing values (drop rows with null income)
df = df.dropna(subset=['Income'])
```

```python
# 2. Remove duplicates
df = df.drop_duplicates()
```

```python
# 3. Standardize text
df['Education'] = df['Education'].str.strip().str.title()
df['Marital_Status'] = df['Marital_Status'].str.strip().str.title()
```

```python
# 5. Rename columns
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
# 6. Create new column: age (optional)
df['age'] = 2025 - df['year_birth']
```

```python
# 7. Check data types
print(df.dtypes)
```

```python
# 8. Save cleaned dataset
df.to_csv("cleaned_marketing_campaign.csv", index=False)
```

```python
# 9.Changing date into datetime for dt customer coloumn
df['dt_customer'] = pd.to_datetime(df['dt_customer'], errors='coerce')

```

```python
df['dt_customer'] = df['dt_customer'].dt.strftime('%d-%m-%Y')
```

```python
# 9.Changing date into datetime for year birth coloumn
df['year_birth'] = pd.to_datetime(df['year_birth'], errors='coerce')
```

```python
df['year_birth'] = df['year_birth'].dt.strftime('%d-%m-%Y')
```

```python
# 11. Check data types
print(df.dtypes)
```

```python
# 12. Save cleaned dataset
df.to_csv("cleaned_marketing_campaign.csv", index=False)
```
